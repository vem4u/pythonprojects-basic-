```python
Making a calculator program!!!

Aritmetic_Operator= input('Please choose an operator: +,-,*,/: ')
num1= float(input('Please choose your first number: '))
num2= float(input('Please choose your next number: '))


if Aritmetic_Operator == '+':
  result= (num1+num2)
  print(round(result, 2))

elif Aritmetic_Operator == '-':
  result= (num1-num2)
  print(round(result, 2))

elif Aritmetic_Operator == '*':
  result= (num1*num2)
  print(round(result, 2))

elif Aritmetic_Operator == '/':
  result= (num1/num2)
  print(round(result, 2))

else:
  print(f"You have not chosen a proper and valid operator!")
```
