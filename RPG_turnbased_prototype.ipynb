```python
#Making a turn based game!!

import random
import time

#variables
player = 100
enemy = 100
atk_low = 15
atk_high = 25

#The intro function
def start():
  time.sleep(1)
  greeting = input('Greetings traveller... \nDo you want to enter the dungeon of doom and despair? (y/n): ').lower()
  if greeting == 'y':
    print('You venture down to your doom...')
    return True
  else:
    print('Someone chickened out...')
    return False

#The combat loop function
def combat():
  global player, enemy, enemy_damage
  turn = 1 #To count the turn
  while player > 0 and enemy > 0:

    print(f'-----Turn {turn} -----')

    time.sleep(1)
    attack = int(input('Press 1 to attack: '))

    if attack == 1:
      time.sleep(1)

      print('Attack registered!')
      time.sleep(1)
      player_damage = random.randint(atk_low, atk_high) #Calculates random damage for player
      enemy -= player_damage #deducts enemy health using the random damage
      enemy = max(0, enemy) #caps enemy health to not go below zero
      time.sleep(1)
      print(f'You have attacked the goblin for {player_damage} dp.\nIts new health is {enemy} hp')

      enemy_damage = random.randint(atk_low, atk_high) #random damage for enemy
      player -= enemy_damage
      player = max(0, player)
      time.sleep(1)

      print(f'The goblin has responded and dealth {enemy_damage} dp to you.\nYour new health is {player} hp. ')
      time.sleep(1)
      if player <= 0 and enemy <0: #Checks if game was a tie
        print('A worthy battle and a glorified death!')
        break
      elif enemy <= 0: #Checks if enemy's health is under zero
       time.sleep(1)
       print('You have slain the goblin, game over!')
       break #breaks loops if true
      elif player <= 0:
       time.sleep(1)
       print('You have been defeated, game over!')
       break
    elif attack != 1: #if user chooses the wrong option, enemy takes advantage
      enemy_damage = random.randint(atk_low, atk_high)
      player -= enemy_damage
      player = max(0, player)
      time.sleep(1)

      print(f'Wrong move! The goblin has taken advantage and dealth {enemy_damage} dp to you.\nYour new health is {player} hp. ')
      if player <= 0: #Checks if the player's health is under zero for second time, if user dies to a wrong move
       time.sleep(1)
      elif player <= 0 and enemy <0: #Same explanation as line#68
        print('A worthy battle and a glorified death!')

        print('You have been defeated, game over!')
        break
    turn += 1 #increments the turn counter everytime a loop finishes

if start(): #start function called
  combat() #combat function starts if start() = y
           #whole game ends if start() = n

#Very fun project to make, took 2 days in total.
#enjoyment= 8/10
#expierience= 10/10
```
