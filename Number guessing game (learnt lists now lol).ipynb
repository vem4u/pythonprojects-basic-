```python
import random

#The Core
low_num = 1
high_num = 100
guesses = 0
guess_his = []
answer = random.randint(low_num, high_num)

#Main game loop
while True:
 user_input = int(input('Take a guess: '))


 user_guess = (user_input)
 guesses += 1
 guess_his.append(user_guess)

 if user_guess < low_num or user_guess > high_num:
    print('Guess out of range!')
    continue
 elif user_guess > answer:
    print('Guess too high!')
    continue
 elif user_guess < answer:
    print('Guess too low!')
    continue
 else:
    print(f'The correct number was {answer} and you took {guesses} tries')
    print(f'Your guesses: {guess_his}')

 #End
 break
```
